```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST

# Define the AlexNet model class
class CustomAlexNet(nn.Module):
    def __init__(self, num_classes=10):
        super(CustomAlexNet, self).__init__()
        self.feature_extractor = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=11, stride=4, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),

            nn.Conv2d(64, 192, kernel_size=5, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),

            nn.Conv2d(192, 384, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),

            nn.Conv2d(384, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),

            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
        )

        # Compute the flattened size dynamically
        dummy_input = torch.zeros(1, 1, 64, 64)  # Single grayscale image of size 64x64
        self.flattened_size = self._compute_flattened_size(dummy_input)

        self.classifier = nn.Sequential(
            nn.Dropout(p=0.5),
            nn.Linear(self.flattened_size, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Linear(4096, num_classes),
        )

    def _compute_flattened_size(self, x):
        x = self.feature_extractor(x)
        return x.view(x.size(0), -1).size(1)

    def forward(self, x):
        x = self.feature_extractor(x)
        x = x.view(x.size(0), -1)  # Flatten the output
        return self.classifier(x)

# Data preparation with transformations
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Resize input images to 64x64
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,)),  # Normalize pixel values to [-1, 1]
])

# Load MNIST dataset
train_dataset = MNIST(root='./data', train=True, download=True, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

test_dataset = MNIST(root='./data', train=False, download=True, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Initialize model, loss function, and optimizer
model = CustomAlexNet(num_classes=10)  # 10 classes for MNIST
loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

def evaluate_model(model, data_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in data_loader:
            outputs = model(inputs)
            _, predictions = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predictions == labels).sum().item()
    return 100 * correct / total

# Training loop
num_epochs = 3
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0
    for batch_idx, (inputs, labels) in enumerate(train_loader):
        optimizer.zero_grad()

        # Forward pass
        outputs = model(inputs)
        loss = loss_function(outputs, labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

        if (batch_idx + 1) % 100 == 0:
            print(f"Epoch [{epoch + 1}/{num_epochs}], Step [{batch_idx + 1}/{len(train_loader)}], Loss: {loss.item():.4f}")

    # Evaluate on the test set after each epoch
    test_accuracy = evaluate_model(model, test_loader)
    print(f"Epoch [{epoch + 1}/{num_epochs}] completed. Loss: {epoch_loss / len(train_loader):.4f}, Test Accuracy: {test_accuracy:.2f}%")

print("Training Complete")

# Final model evaluation
final_accuracy = evaluate_model(model, test_loader)
print(f"Final Test Accuracy: {final_accuracy:.2f}%")
```

```python
import torch
import torch.nn as nn

class CustomAlexNet(nn.Module):
    def __init__(self, num_classes=100):
        super(CustomAlexNet, self).__init__()
        self.features = nn.Sequential(
            # Conv1 Layer: 3 input channels, 32 output channels, 7x7 kernel, stride 2, padding 3
            nn.Conv2d(3, 32, kernel_size=7, stride=2, padding=3),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),

            # Conv2 Layer: 32 input channels, 64 output channels, 5x5 kernel, stride 1, padding 2
            nn.Conv2d(32, 64, kernel_size=5, stride=1, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),

            # Conv3 Layer: 64 input channels, 128 output channels, 3x3 kernel, padding 1
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),

            # Conv4 Layer: 128 input channels, 128 output channels, 3x3 kernel, padding 1
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),

            # Conv5 Layer: 128 input channels, 128 output channels, 3x3 kernel, padding 1
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
        )

        # Adjusted fully connected layers
        self.classifier = nn.Sequential(
            nn.Dropout(),
            nn.Linear(128 * 7 * 7, 2048),  # Flattened size adjusted
            nn.ReLU(inplace=True),
            nn.Dropout(),
            nn.Linear(2048, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.classifier(x)
        return x


# Example usage of the model
model = CustomAlexNet(num_classes=100)  # Changed num_classes to 100 for variety
input_tensor = torch.randn(1, 3, 128, 128)
output = model(input_tensor)
from torchsummary import summary
print("Output shape:", output.shape)
summary(model, ( 3, 128, 128))
```
